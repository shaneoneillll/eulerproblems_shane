```python
total_sum = 0
for number in range(0,1001):
    #split up number into components
    sub_deci = number%10
    deci = number%100 - sub_deci%10
    hunned = number - deci - sub_deci
    hunned_index = str(hunned)
    
    #account for teen values
    teen = 0
    if deci == 10:
        teen = deci + sub_deci
        deci = 0
        sub_deci = 0
    
    #define numbers in their letter strings and count the letters in the string
    letter_sum = 0
    if hunned != 0:
        a = numbers[hunned_index[0]] + "hundredand" + numbers[str(teen)] + numbers[str(deci)] + numbers[str(sub_deci)]
        if int(str(number)[1:3]) == 0:
            a = numbers[hunned_index[0]] + "hundred"
        if number == 1000:
            a = "onethousand" 
        print(a)
        for letter in a:
            letter_sum += 1
            
    else:
        b = numbers[str(teen)] + numbers[str(deci)] + numbers[str(sub_deci)]
        print(b)
        for letter in b:
            letter_sum += 1
        
    
        
    
    total_sum += letter_sum 
total_sum
```
