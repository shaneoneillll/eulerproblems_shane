```python
y = 0
for i in range(100,1000): #loop counts from 100 to 999 to cycle through all 3-digit numbers
    j = 0
    for j in range(100,1000): #loopception
        x = i*j #multiply both numbers together
        
        pal = str(x) #make that number a lst of seperate strings for each digit
        pal_rev = pal[::-1] # reverse that list
        if pal == pal_rev: #if they are the same:
            if int("".join(pal_rev)) > int(y): #converts list of strings to one string and then their numerical value
                y = pal #let y hold the new highest palindrome
            print(i, j, x, y)
           
        
```
