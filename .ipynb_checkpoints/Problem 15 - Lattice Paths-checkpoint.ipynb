```python
#Recognise that a 20x20 list-matrix will simulate the grid in boxes but not in latices
#Need a grid of 21X21 to actully simulate the latices. Think about a 2x2 grid. 4 boxes made by 9 latices

def path_counter(x):
    y = x + 1
    first_row = []
    for i in range (0,y):
        first_row.append(1)

    paths_at_point = [first_row]

    for row in range (0,y):
        new_row = [1]
        for column in range (0,(x)):
            new_row.append(new_row[column]+paths_at_point[row][column+1])

        paths_at_point.append(new_row)
    return(paths_at_point)
```

```python
path_counter(20)
```
